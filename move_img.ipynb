```python
import os
import shutil
import json

# Percorsi delle cartelle e del file JSONL
source_folder = r"C:\Users\utente\Desktop\Università\Themis-SEMEVAL-public\images\2b"
destination_folder = r"C:\Users\utente\Desktop\Università\Themis-SEMEVAL-public\images\dev"
jsonl_file_path = r"C:\Users\utente\Desktop\Università\Themis-SEMEVAL-public\annotations\subtask2b\dev_unlabeled.jsonl"

# Crea la cartella di destinazione se non esiste
os.makedirs(destination_folder, exist_ok=True)

# Leggi il file JSONL e sposta le immagini specificate
with open(jsonl_file_path, 'r') as file:
    for line in file:
        entry = json.loads(line.strip())
        image_name = entry["img"]
        source_image_path = os.path.join(source_folder, image_name)
        destination_image_path = os.path.join(destination_folder, image_name)
        
        if os.path.exists(source_image_path):
            shutil.move(source_image_path, destination_image_path)
            print(f"Spostato: {image_name}")
        else:
            print(f"Immagine non trovata: {image_name}")
```
